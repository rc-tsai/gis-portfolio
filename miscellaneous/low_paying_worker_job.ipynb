```python
import pandas as pd
import os
```

```python
os.chdir("D:\M.S. GIS\Georgia Tech\GRA\CQGRD\Job_Accessibility")
df = pd.read_csv("ga_od_main_JT01_2021.csv")
df
```

#### Obtain total low-paying workers at census tract

```python
# select columns
df_low_worker = df[['h_geocode','SE01']].copy()

# aggregate workers from block to census tract level
df_low_worker['CT_GEOID'] = df_low_worker['h_geocode'].astype(str).str[:11]
df_low_worker['low_worker'] = df_low_worker.groupby('CT_GEOID')['SE01'].transform('sum')
df_low_worker = df_low_worker.drop_duplicates(subset=['CT_GEOID'])

# check if the workers in the subset add up correctly
print("------------------------------------------")
print(df_low_worker['low_worker'].sum() == df['SE01'].sum())

df_low_worker = df_low_worker.reset_index()
df_low_worker = df_low_worker[['CT_GEOID', 'low_worker']]
df_low_worker
```

#### Obtain total low-paying jobs at census tract

```python
# select columns
df_low_job = df[['w_geocode','SE01']].copy()

# aggregate jobs from block to census tract level
df_low_job['CT_GEOID'] = df_low_job['w_geocode'].astype(str).str[:11]
df_low_job['low_job'] = df_low_job.groupby('CT_GEOID')['SE01'].transform('sum')
df_low_job = df_low_job.drop_duplicates(subset=['CT_GEOID'])

# check if the jobs in the subset add up correctly
print("------------------------------------------")
print(df_low_job['low_job'].sum() == df['SE01'].sum())

df_low_job = df_low_job.reset_index()
df_low_job = df_low_job[['CT_GEOID', 'low_job']]
df_low_job
```
